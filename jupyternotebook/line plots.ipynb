# Draw a line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
i=sns.load_dataset("iris")

#draw a line plot
sns.lineplot(x="species", y="petal_width", data=i)
plt.show()
```

# Adding heading of the line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
i=sns.load_dataset("iris")

#draw a line plot
sns.lineplot(x="species", y="petal_width", data=i)

#add title of plot
plt.title("Line graph for the specie and petal width of the data set iris")
plt.show()
```

# Adding limit to the x-axis or y-axis

```python
import seaborn as sns
import matplotlib.pyplot as plt
i=sns.load_dataset("iris")

#draw a line plot
sns.lineplot(x="species", y="petal_width", data=i)

#Adding limit to the y-axis because x-axis is not numeric
plt.ylim(1)

#add title of plot
plt.title("Line graph for the specie and petal width of the data set iris")
plt.show()
```

# Change the styles
- Dark grid
- White grid
- White
- ticks
- Dark

```python
import seaborn as sns
import matplotlib.pyplot as plt
i=sns.load_dataset("iris")

#draw a line plot
sns.lineplot(x="species", y="petal_width", data=i)

#Adding limit to the y-axis because x-axis is not numeric
plt.ylim(1)

#Adding the syle of graph
sns.set(style="dark")

#add title of plot
plt.title("Line graph for the specie and petal width of the data set iris")
plt.show()
```
